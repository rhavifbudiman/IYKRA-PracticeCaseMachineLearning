## MARKET BASKET ANALYSIS USING APRIORI


```python
#grouping data based on purposes
basket_uk = (data[data['Country'] =='United Kingdom'].groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo')) 
basket_uk.head()
```

```python
#transform data for modelling
def hot_encode(x): 
    if(x<= 0): 
        return 0
    if(x>= 1): 
        return 1
  
# Encoding the datasets 
basket_encoded = basket_uk.applymap(hot_encode) 
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
# Building the mode
frq_items = apriori(basket_encoded, min_support = 0.03, use_colnames = True) 
  
frq_items.head(10)
```

```python
rules = association_rules(frq_items, metric="lift", min_threshold=0.5)
rules.head()
```

so from this analysis, we can conclude if we buy ALARM CLOCK BAKELIKE GREEN many people also buy ALARM CLOCK BAKELIKE RED
